## Model Building
#### Splitting Data into Training Sets and Test Sets

```python
# splitting the data
X= df_iris.drop(["Id","Species"],axis=1)
y=df_iris['Species']
```

```python
# train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(X_train,y_train)
```

```python
Y_pred = dtc.predict(X_test)
```

```python
df2 = pd.DataFrame({"Actual":y_test,"Predicted":Y_pred})
df2.head()
```

```python
# Testing the accuracy of model prediction
accuracy_score(y_test,Y_pred)
```

```python
new_model= DecisionTreeClassifier().fit(X,y)
```

```python
feature_name=["Sepal length", "Sepal width", "Petal length", "Petal width"]
class_name=["Setosa", "Versicolor", "Virginica"]
```

### Plotting Decision Tree

```python
plt.figure(figsize = (15,10))
tree = tree.plot_tree(new_model, feature_names = feature_name, class_names = class_name, filled = True)
```
